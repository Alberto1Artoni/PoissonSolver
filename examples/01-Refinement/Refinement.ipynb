# **Refinement**

Code to visualize the refinement of a given structured grid.

```python
# import packages
from PoissonSolver import mesh
import json
import matplotlib.pyplot as plt
```

```python
# read data
filePath = "data.json"
with open(filePath, 'r') as file:
    data = json.load(file)
```

```python
# generate mesh
grid = mesh.Mesh(data)
# initial grid
grid.plot()
```

```python
# Loop and refine
for i in range(0,3):
    # start with a finer refinement
    grid.refine()
    grid.plot()
```
